1.Which country has the most number of players (score :1)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ntbl=pd.read_csv('fifa_data.csv')
ntbl
```

```python
new1=pd.read_csv('fifa_data.csv',usecols=['Name','Nationality'])
new1
```

```python
new1.drop_duplicates(inplace=True)
new1
```

```python
new = new1.groupby(['Nationality']).size().reset_index(name='Number of people').sort_values(['Number of people'])
display(new.tail(1))
```

Insight :England has the highest no of players with 1569 distinct players

2.Plot a bar chart of 5 top countries with the most number of players. (score :1)

```python
new2=new.tail(5)
new2
```

```python
x=list(new2.iloc[:,0])
y=list(new2.iloc[:,1])
plt.bar(x,y,color='g')
plt.title("Countries with largest no of players")
plt.xlabel("Countries")
plt.ylabel("No. of players")
plt.show()
```
